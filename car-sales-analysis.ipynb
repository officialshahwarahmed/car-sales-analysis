```python
import pandas as pd
```

```python
cars = pd.read_csv("cars.csv", encoding = "Latin-1")
```

```python
cars
```

```python
cars["brand"].value_counts()
```

```python
selling_prop = cars["brand"].value_counts(normalize = True)
selling_prop
```

```python
selling_bol = selling_prop>0.05
selling_bol
```

```python
selling_prop[selling_bol]
```

```python
top_brand = selling_prop[selling_bol]
top_brand
```

```python
brand_name = top_brand.index
```

```python
brand_name
```

```python
brand_bol = cars["brand"].isin(brand_name)
brand_bol
```

```python
cars.head()
```

```python
top_selling_brand = cars[brand_bol]
top_selling_brand
```

```python
top_selling_brand["price"]
top_selling_brand["price"] = top_selling_brand["price"].str.replace("$","").str.replace(",","").astype(int)
top_selling_brand["price"]
```

```python
top_selling_brand["odometer"] = top_selling_brand["odometer"].str.replace("km","").str.replace(",","").astype(int)
top_selling_brand["odometer"]
```

```python
top_selling_brand.rename({"price":"Price ($)","odometer":"Odometer (km)"},axis = 1, inplace = True)
```

```python
top_selling_brand
```

```python
grouping = top_selling_brand.groupby("brand")
```

```python
grouping[["Price ($)","Odometer (km)"]].mean()
```

```python
brand_mean = grouping[["Price ($)","Odometer (km)"]].mean()
```

```python
brand_mean.sort_values("Price ($)",ascending = False)
```
